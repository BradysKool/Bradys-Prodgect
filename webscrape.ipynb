```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import Main
```

```python
url = 'https://www.dell.com/en-us/shop/dell-laptops/sr/laptops/alienware-laptops'
page = requests.get(url)

# Response 200 good to go
page = page.text

soup = BeautifulSoup(page, 'html.parser')

title = soup.title.text
mainPage = soup.find_all('article')
```

```python
pc_titles = Main.findtitles(mainPage)

def data_clean(pc_titles):
    clean_titles = []
    for t in pc_titles:
        t = t.replace('\n','')
        clean_titles.append(t)
    return  clean_titles

title = data_clean(pc_titles)
title
```

```python
pc_price = Main.find_prices(mainPage)
pc_price
def data_clean(pc_price):
    clean_titles = []
    for t in pc_price:
        clean_titles.append(t)
    return  clean_titles


p = []
for i in range (len(pc_price)):
    if pc_price[i] == None:
        break
    price = (pc_price[i]).find(class_ = None)
    #price =  price.replace('<span>', ' ')
    p.append(price)

price_clean = []





```

```python
def get_table_data(mainPage):
    CPU = [] 
    Operating_system = []
    GPU = []
    Memory = []
    Hard_drive = []
    Display = []
    Weight = []
    for artical in mainPage:
        info = artical.find('div', class_ ='ps-short-specs ps-iconography')
        if (info == None):
            break
        specs = info.find_all('div')
        CPU.append(specs[0])
        Operating_system.append(specs[1])
        GPU.append(specs[2])
        Memory.append(specs[3])
        Hard_drive.append(specs[0])
    for artical in mainPage:
        otherinfo = artical.find('div', class_='ps-features-icon')
        if (otherinfo == None):
            break
        dainfo = otherinfo.find_all('div')
        Display.append(dainfo[0])
        Weight.append(dainfo[1])
    return CPU, Operating_system, GPU, Memory, Display, Weight
CPU, Operating_system, GPU, Memory, Display, Weight = get_table_data(mainPage)

def clear_whitespace(values):
    clear = []
    for val in values:
        val_text = val.text
        #Gets rid of white space
        clean_space = val_text.replace(' ','')
        #
        clean_n = clean_space.replace('\n', '')
        clear.append(clean_n)
    return clear

Cpu = clear_whitespace(CPU)
Operating_system = clear_whitespace(Operating_system)
GPU = clear_whitespace(GPU)
Memory = clear_whitespace(Memory)
Display = clear_whitespace(Display)
Weight = clear_whitespace(Weight)


#Cleans Display
def Display_Clean(Display):
    num = []
    for number in Display:
        just_number = number.replace('-in.display','')
        num.append(just_number)
    return num
Display = Display_Clean(Display)


#cleans Weight
def Weight_Clean(Weight):
    lbs = []
    for number in Weight:
        number = number[11:15]
        lbs.append(number)
    return lbs
Weight = Weight_Clean(Weight)
Weight



        

```
